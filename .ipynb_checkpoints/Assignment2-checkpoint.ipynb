```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import sklearn
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

# Task 1

```python
df.info()
```

```python
#Data Analysis
df.isna().sum()
```

```python
#Fill all the numeric rows with median
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            # Fill missing numeric values with median
            df[label] = content.fillna(content.median())
```

```python
df.isna().sum()
```

```python
#Removing the all rows of remaining null values in 3 columns
df = df[~df['Mileage'].isna()]
```

```python
df = df[~df['New_Price'].isna()]
df = df[~df['Engine'].isna()]
df = df[~df['Power'].isna()]
```

```python
df.isna().sum()
```

# Task2

```python
df.head()
```

```python
#Removing the unit for the rows having
df['Mileage'] = df['Mileage'].str.split(' ').str.get(0)
df.head()
```

```python
#Remove the units for all the other required columns
df['Engine'] = df['Engine'].str.split(' ').str.get(0)
df['Power'] = df['Power'].str.split(' ').str.get(0)
df['New_Price'] = df['New_Price'].str.split(' ').str.get(0)
df.head()
```

# Task 3

```python
#Now 

from sklearn.preprocessing import OneHotEncoder 

df['Fuel_Type'] = df['Fuel_Type'].astype('category')
df['Transmission'] = df['Transmission'].astype('category')

# Assigning numerical values and storing it in another columns 
df['Fuel_type_new'] = df['Fuel_Type'].cat.codes 
df['Transmission_new'] = df['Transmission'].cat.codes

# Create an instance of One-hot-encoder 
encoding = OneHotEncoder() 
  
# Passing encoded columns 
  
encoding_data = pd.DataFrame(encoding.fit_transform(df[['Fuel_type_new', 'Transmission_new']]).toarray()) 
  
# Merge with main Dataset
New_df = df.join(encoding_data) 
  
print(encoding_data) 
```

```python
#Printing the new data frame with the encoding values
New_df.head()
```

# Task 4

```python
#Now we need to add a column to dataset

from datetime import datetime

current_year = datetime.now().year
df['Car_Current_Age'] = current_year - df['Year']
```
